```python
"""
TextLoaderä¼šå°†ä¸€ä¸ªæ–‡æ¡£æ–‡ä»¶åŠ è½½ä¸ºä¸€ä¸ªDocumentå¯¹è±¡ï¼Œè¯¥å¯¹è±¡æœ‰ä¸¤ä¸ªå±æ€§ï¼š
    metadata:       å­˜å‚¨è¯¥æ–‡æ¡£çš„æ¥æºè·¯å¾„ç­‰å…ƒæ•°æ®
    page_content:   å­˜å‚¨æ–‡æ¡£çš„å†…å®¹
"""

from langchain_community.document_loaders import TextLoader

text_documents = TextLoader("knowledge_base/sample.txt", encoding="utf-8").load()
print(text_documents)
```

```python
"""
UnstructuredMarkdownLoaderå¯ç”¨äºåŠ è½½Markdownæ–‡ä»¶
    mode: åŠ è½½æ¨¡å¼
        "single"    è¿”å›å•ä¸ªDocumentå¯¹è±¡
        "elements"  æŒ‰æ ‡é¢˜ç­‰å…ƒç´ åˆ‡åˆ†æ–‡æ¡£
    strategy: åŠ è½½ç­–ç•¥
        "fast"      å¿«é€Ÿç²—ç²’åº¦åŠ è½½
        "hi_res"    ç»†ç²’åº¦åŠ è½½ï¼ŒæŒ‰æ ‡é¢˜å±‚çº§ã€åˆ—è¡¨é¡¹ã€è¡¨æ ¼ç­‰ç»“æ„ç»†åˆ†
"""

from langchain_community.document_loaders import UnstructuredMarkdownLoader

md_documents = UnstructuredMarkdownLoader(
    "knowledge_base/sample.md",
    mode="elements",
    strategy="fast",
).load()
print(md_documents)
```

```python
"""
PyPDFLoader
    æ”¯æŒé¡µçº§æ‹†åˆ†ï¼Œæ¯ä¸€é¡µä½œä¸ºä¸€ä¸ªDocumentè¿”å›
    æ”¯æŒæå–å›¾åƒã€æå–å¸ƒå±€
    extraction_mode: æå–æ¨¡å¼
        "plain"     æå–çº¯æ–‡æœ¬
        "layout"    æå–å¸ƒå±€
"""

from langchain_community.document_loaders import PyPDFLoader

pdf_documents = PyPDFLoader(
    "knowledge_base/sample.pdf",
    extraction_mode="layout",
).load()
print(pdf_documents)
```

```python
"""
UnstructuredPDFLoader
    æ”¯æŒç»“æ„åŒ–æå–ï¼Œæ”¯æŒOCR
    ä»…å½“ PDF æ–‡æ¡£ä¸­ä¸å­˜åœ¨æ–‡æœ¬æ—¶ï¼Œæ‰ä¼šåº”ç”¨ OCR
    mode: åŠ è½½æ¨¡å¼
        "single"    è¿”å›å•ä¸ªDocumentå¯¹è±¡
        "elements"  æŒ‰æ ‡é¢˜ç­‰å…ƒç´ åˆ‡åˆ†æ–‡æ¡£
    strategy: åŠ è½½ç­–ç•¥
        "fast"      æå–å¹¶å¤„ç†æ–‡æœ¬
        "ocr_only"  å…ˆè¿›è¡Œ OCR å¤„ç†ï¼Œå†å¤„ç†åŸå§‹æ–‡æœ¬
        "hi_res"    è¯†åˆ«æ–‡æ¡£å¸ƒå±€å¹¶å¤„ç†ï¼Œè‡ªåŠ¨ä¸‹è½½YOLOXæ¨¡å‹ï¼ˆè¯†åˆ«é¡µé¢å¸ƒå±€ï¼‰
    infer_table_structure: æ˜¯å¦æ¨æ–­è¡¨æ ¼ç»“æ„
        ä»… hi_res ä¸‹èµ·æ•ˆ
        å¦‚æœä¸º Trueï¼Œæå–å‡ºçš„è¡¨æ ¼å…ƒç´ ä¼šåŒ…å«ä¸€ä¸ª text_as_html å…ƒæ•°æ®ï¼Œå°†æ–‡æœ¬å†…å®¹è½¬æ¢ä¸º html æ ¼å¼
    languages: OCRä½¿ç”¨çš„è¯­è¨€
        éœ€ä¼ å…¥è¯­è¨€åˆ—è¡¨
        è¯­è¨€åˆ—è¡¨å‚è€ƒ https://github.com/tesseract-ocr/langdata
    æ›´å¤šå‚æ•°è¯¦è§ https://github.com/Unstructured-IO/unstructured/blob/main/unstructured/partition/pdf.py
"""

from langchain_community.document_loaders import UnstructuredPDFLoader

pdf_documents = UnstructuredPDFLoader(
    "knowledge_base/sample.pdf",
    mode="elements",
    strategy="hi_res",
    infer_table_structure=True,
    languages=["eng", "chi_sim"],
).load()
print(pdf_documents)
```

```python
"""
UnstructuredWordDocumentLoader
    é€‚ç”¨äº .docx å’Œ .doc æ–‡ä»¶
    mode: åŠ è½½æ¨¡å¼
        "single"    è¿”å›å•ä¸ªDocumentå¯¹è±¡
        "elements"  æŒ‰æ ‡é¢˜ç­‰å…ƒç´ åˆ‡åˆ†æ–‡æ¡£
    strategy: åŠ è½½ç­–ç•¥
        "fast"      å¿«é€Ÿç²—ç²’åº¦åŠ è½½
        "hi_res"    ç»†ç²’åº¦åŠ è½½ï¼ŒæŒ‰ç»“æ„ç»†åˆ†
"""

from langchain_community.document_loaders import UnstructuredWordDocumentLoader

word_documents = UnstructuredWordDocumentLoader(
    "knowledge_base/sample.docx",
    mode="elements",
    strategy="fast",
).load()
print(word_documents)
```

```python
"""
WebBaseLoader
    é€‚ç”¨äºç½‘é¡µ
    web_paths: ç½‘å€åºåˆ—
    bs_kwargs: ä¼ ç»™ BeautifulSoup çš„è§£æå‚æ•°
        parse_only  åªæå–æŒ‡å®šæ ‡ç­¾çš„å…ƒç´ 
"""
import bs4
from langchain_community.document_loaders import WebBaseLoader

web_documents = WebBaseLoader(
    web_paths=(
        " https://news.sina.com.cn/c/xl/2025-09-07/doc-infprmwn0510979.shtml",
    ),
    bs_kwargs={"parse_only": bs4.SoupStrainer(id="article")},  # åªæå–æ­£æ–‡åŒºåŸŸ
).load()
print(web_documents)
```
